```python
#pre-processing the images

def pre_process_image(image_path, compression_factor=2):
    # Read the image
    image = cv2.imread(image_path)

    # Convert to grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Access the grayscale image as a matrix
    grayscale_matrix = np.array(grayscale_image)

    
    # Compress the image by resizing
    original_height, original_width = grayscale_image.shape
    new_height = original_height // compression_factor
    new_width = original_width // compression_factor

    compressed_image = cv2.resize(grayscale_image, (new_width, new_height), interpolation=cv2.INTER_AREA)
    compressed_matrix = np.array(compressed_image)

    grayscale_matrix = grayscale_matrix / 255.0
    compressed_matrix = compressed_matrix / 255.0
    return grayscale_matrix, compressed_matrix

def display_image(image_matrix, title="Image"):
    cv2.imshow(title, image_matrix)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def save_image(image_matrix, file_name):
    cv2.imwrite(file_name, image_matrix)
    print(f"Image saved as '{file_name}'")
```

```python
# Loading the model and predicting from image
def load_nn_model(model_path):
    model = load_model(model_path)
    return model

def predict_from_image(model, image_matrix, window_size=5): # Uses sliding window to process the entire image in batches
    height, width = image_matrix.shape
    predictions = []
    row_limit = height
    col_limit = width
    print(f"Making predictions for {row_limit} rows and {col_limit} columns")
    # Slide the window across the image
    for i in range(0, row_limit, window_size):
        print(f"Predicting for {i}/{row_limit}")
        for j in range(0, col_limit, window_size):
            # Extract the patch
            patch = image_matrix[i:i + window_size, j:j + window_size]
            # Preprocess the patch (flatten and reshape for Conv1D model)
            patch = patch.flatten()
            # print(f"Predicting for {i}/{row_limit}, {j}/{col_limit}")
            patch = patch.reshape(1, -1, 1)
            # Make predictions
            pred = model.predict(patch)

            # Interpret predictions
            has_edge = "Yes" if pred[0][0] > 0.5 else "No"  # Binary output (threshold at 0.5)
            edge_no1 = np.argmax(pred[1], axis=1)[0]
            edge_no2 = np.argmax(pred[2], axis=1)[0]
            if has_edge == "Yes" and edge_no1 > 0 and edge_no2 > 0:
                if window_size == 3:
                    edge_matrix = matrices3x3[f'mat_3x3_{edge_no1}{edge_no2}']
                elif window_size == 5:
                    edge_matrix = matrices5x5[f'mat_5x5_{edge_no1}{edge_no2}']
                else:
                    edge_matrix = matrices7x7[f'mat_7x7_{edge_no1}{edge_no2}']
            else:
                edge_matrix = np.zeros((window_size, window_size))

            # Replace the corresponding part of the image matrix with the edge matrix
            image_matrix[i:i + window_size, j:j + window_size] = edge_matrix

            predictions.append((has_edge, edge_no1, edge_no2))
            

    return predictions, image_matrix

def make_edge_matrix(image_matrix, predictions, window_size):
    height, width = image_matrix.shape
    row_limit = height
    col_limit = width
    output_image = np.copy(image_matrix)
    x = 0
    for i in range(0, row_limit, window_size):
        for j in range(0, col_limit, window_size):
            
            has_edge, edge_no1, edge_no2 = predictions[x]
            x+=1
            if has_edge == "Yes" and edge_no1 > 0 and edge_no2 > 0:
                # Determine the matrix size based on the edge_no1
                if window_size == 3:
                    edge_matrix = matrices3x3[f'mat_3x3_{edge_no1}{edge_no2}']
                elif window_size == 5:
                    edge_matrix = matrices5x5[f'mat_5x5_{edge_no1}{edge_no2}']
                else:
                    edge_matrix = matrices7x7[f'mat_7x7_{edge_no1}{edge_no2}']
            else:
                edge_matrix = np.zeros((window_size, window_size))

            # Replace the corresponding part of the image matrix with the edge matrix
            output_image[i:i + window_size, j:j + window_size] = edge_matrix

    return output_image

def neural_newtork_edge_detection(image_path, model_path, output_path, compression_factor=2, window_size=5):
    
    model = load_model(model_path)
    print(f"Model loaded from {model_path}")
    grayscale_matrix, compressed_matrix = pre_process_image(image_path, compression_factor)
    print(f"Image pre-processed with compression factor {compression_factor}")
    print(f"Compressed image shape: {compressed_matrix.shape}")
    predictions, image_matrix = predict_from_image(model, compressed_matrix, window_size)
    print(f"Predictions made for {len(predictions)} patches")
    return predictions, image_matrix
    # # output_matrix = make_edge_matrix(compressed_matrix, predictions, window_size)
    # output_matrix = image_matrix
    # print(f"Edge matrices made for {len(predictions)} patches")
    # save_image(output_matrix, output_path)
    # print(f"Output image saved to {output_path}")
    # display_image(output_matrix, title="Output Image")


```
